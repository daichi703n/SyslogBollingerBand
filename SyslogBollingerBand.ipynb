```python
import pandas as pd
df = pd.read_csv('/var/log/system.log', delim_whitespace=True, header=None, error_bad_lines=False)
count = df.groupby(pd.TimeGrouper('1Min')).count()
mean = count.rolling(window=60).mean()
std = count.rolling(window=60).std()
std_plus = std.apply(lambda  x: x * 2)
std_minus = std.apply(lambda x: x * -2)
upper_limit = mean.add(std_plus)
lower_limit = mean.add(std_minus)
```
